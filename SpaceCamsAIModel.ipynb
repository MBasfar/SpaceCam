```python
import os

label_dir = '/content/drive/MyDrive/dataset/Train/labels'  # Update this path as needed

# Define a mapping to correct the labels, assuming you want to map label 15 to a correct class (e.g., 2)
label_mapping = {
    15: 0,
    16: 1,
    17: 2# Change '15' to class '2' or any other appropriate class
}

# Iterate over all label files and fix them
for label_file in os.listdir(label_dir):
    if label_file.endswith('.txt'):
        file_path = os.path.join(label_dir, label_file)
        corrected_lines = []
        with open(file_path, 'r') as f:
            lines = f.readlines()
            for line in lines:
                parts = line.strip().split()
                class_id = int(parts[0])
                if class_id in label_mapping:
                    parts[0] = str(label_mapping[class_id])
                corrected_lines.append(" ".join(parts))
        # Save the corrected annotations back to the file
        with open(file_path, 'w') as f:
            f.write("\n".join(corrected_lines))
```
